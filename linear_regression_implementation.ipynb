# Implementation of Simple Linear Regression

```python
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
class Raw_Linear_Regression:
    def __init__(self):
        self.m = None
        self.b = None

    def train(self, X_train, Y_train):
        num = 0
        den = 0
        X_mean = X_train.mean()
        Y_mean = Y_train.mean()

        for i in range(X_train.shape[0]):
            num += (Y_train[i] - Y_mean)*(X_train[i] - X_mean)
            den += (X_train[i] - X_mean)*((X_train[i] - X_mean))
        
        self.m = num/den
        self.b = Y_mean - self.m*X_mean
        print(self.m)
        print(self.b)

    def predict(self,X_test):
        return self.m*X_test + self.b
```

```python
df = pd.read_excel('C:\D\Codes\Machine_Learning\student_cgpa_package_dataset.xlsx')
df.head()
```

```python
X = df.iloc[:,0].values
Y = df.iloc[:,1].values
```

```python
X_train,X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
X_train.shape
```

```python
model = Raw_Linear_Regression()

```

```python
model.train(X_train, Y_train)
```

```python
lst = model.predict(X_test)
print(lst[0:10])
```
