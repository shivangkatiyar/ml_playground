# Simple Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_excel('C:/D/Codes/Machine_Learning/student_cgpa_package_dataset.xlsx')
df.columns
```

```python
plt.scatter(df['package_lpa'], df['cgpa'])
plt.xlabel('cgpa')
plt.ylabel('package_lpa')
plt.show()
```

```python
X = df[['cgpa']]
Y = df['package_lpa']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

### Now with model.fit(X_train, Y_train)

model will try to create a best fit line or we can say it will find the best value of slope(m) and y-intercept(b or c)

package = m * cgpa + b

```python
model.fit(X_train, Y_train)
```

```python
print(model.coef_)  # value of slope (m)
print(model.intercept_) # value of intercept (c or b)
```

```python
Y_pred = model.predict(X_test)
```

```python
comparison = X_test.copy()
comparison['Actual'] = Y_test.values
comparison['Predicted'] = Y_pred


comparison.head()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print('MAE', mean_absolute_error(Y_test, Y_pred))
print('MSE', mean_squared_error(Y_test, Y_pred))
print('MAE', r2_score(Y_test, Y_pred))
```
